```python
# importing packages
import seaborn as sns
import matplotlib.pyplot as plt

#Visualizing the target column
sns.countplot(x='Species', data=df, )
plt.show()


```

```python
#Relation between variables
sns.scatterplot(x='Culmen Length (mm)', y='Culmen Depth (mm)',
				hue='Species', data=df, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
# Comparing Petal Length and Petal Width
sns.scatterplot(x='Culmen Length (mm)', y='Culmen Depth (mm)',
				hue='Species', data=df, )

# Placing Legend outside the Figure
plt.legend(bbox_to_anchor=(1, 1), loc=2)

plt.show()
```

```python
sns.pairplot(df.drop(['Individual ID'], axis = 1),
			hue='Species', height=2)
```

```python
#distribution of data for various columns

fig, axes = plt.subplots(2, 2, figsize=(10,10))

axes[0,0].set_title("Culmen Length")
axes[0,0].hist(df['Culmen Length (mm)'], bins=7)

axes[0,1].set_title("Culmen Depth")
axes[0,1].hist(df['Culmen Depth (mm)'], bins=5);

axes[1,0].set_title("Flipper Length")
axes[1,0].hist(df['Flipper Length (mm)'], bins=6);

axes[1,1].set_title("Body Mass")
axes[1,1].hist(df['Body Mass (g)'], bins=6);
```

```python
 #importing packages
import seaborn as sns
import matplotlib.pyplot as plt

plot = sns.FacetGrid(df, hue="Species")
plot.map(sns.distplot, "Culmen Length (mm)").add_legend()

plot = sns.FacetGrid(df, hue="Species")
plot.map(sns.distplot, "Culmen Depth (mm)").add_legend()

plot = sns.FacetGrid(df, hue="Species")
plot.map(sns.distplot, "Flipper Length (mm)").add_legend()

plot = sns.FacetGrid(df, hue="Species")
plot.map(sns.distplot, "Body Mass (g)").add_legend()

plt.show()
```

```python
#boxplots to see how the categorical value is distributed with other numerical values

def graph(y):
	sns.boxplot(x="Species", y=y, data=df)

plt.figure(figsize=(10,10))

# Adding the subplot at the specified
# grid position
plt.subplot(221)
graph('Culmen Length (mm)')

plt.subplot(222)
graph('Culmen Depth (mm)')

plt.subplot(223)
graph('Flipper Length (mm)')

plt.subplot(224)
graph('Body Mass (g)')

plt.show()
```
